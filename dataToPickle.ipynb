```python
path="/gdrive/My Drive/Colab Notebooks/dacon/northPole"
files=os.listdir(path+"/train")

train_x_size=len(files)
files.sort()

meltL=[]
missingL=[]
for fileName in files:
    my_npy=np.load(path+f"/train/{fileName}")
    
    meltL.append(my_npy[:,:,0])
    missingL.append(my_npy[:,:,0])

    # 데이터시각화
    # show_data(npy=my_npy) 

meltL=np.array(meltL)
missingL=np.array(missingL)

# 피클데이터로
with open(path+'/dataP/melt.pickle','wb') as fw:
    pickle.dump(meltL, fw)

with open(path+'/dataP/missing.pickle','wb') as fw:
    pickle.dump(missingL, fw)
```
